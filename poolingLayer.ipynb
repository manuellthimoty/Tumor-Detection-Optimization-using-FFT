```python
import numpy as np
```

```python
class PoolingLayer():
    def _init_(self,pool_size = 2, stride = 2):
        self.pool_size = pool_size
        self.stride = stride

    def forward(self,input):
        self.input = input
        batch_size, input_channel, input_h,input_w = input.shape
        out_h = int((input_h - self.pool_size) / self.stride) + 1
        out_w = int((input_w - self.pool_size) / self.stride) + 1
        output = np.zeros((batch_size, input_channel,out_h,out_w))

        for b in range (batch_size):
            for c in range (input_channel):
                for i in range (out_h) :
                    for j in range(out_w):
                        hStart = i * self.stride
                        hEnd = hStart + self.pool_size
                        wStart = j * self.stride
                        wEnd = hStart + self.pool_size
                        cur = input[b,c,hStart:hEnd,wStart:wEnd]

                        output[b,c,i,j] = np.max(cur)
        return output
```
